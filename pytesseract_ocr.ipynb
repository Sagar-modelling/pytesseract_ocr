```python
import pytesseract
import shutil
import os
import numpy as np
import random
try:
 from PIL import Image
except ImportError:
 import Image
```

```python
# Get bounding box estimates
print(pytesseract.image_to_boxes(img))
```

```python
extractedInformation = pytesseract.image_to_string(img)
print(extractedInformation)
```

```python
with open('structured_output.txt', mode = 'w') as f:  #create a file conataining all extracted text information
    f.write(extractedInformation)
```

```python
extracted_data = {}
with open('structured_output.txt', 'r') as text_file: #read the text file
    line_lst = text_file.readlines()
line_lst
```

```python
for line in line_lst:
    line_strip = line.strip()
    if 'Name' in line_strip:
        extracted_data['Name'] = " ".join(line_strip.split()[1:])
    if 'DOB' in line_strip:
        extracted_data['Date of Birth'] = line_strip.split()[1:]
```

```python
print(extracted_data)
```

```python
img = cv2.imread('/content/text_demof1.jpg')
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
ret, thresh = cv2.threshold(img_gray, 70, 255, cv2.THRESH_BINARY)
```

```python
extractedInformation = pytesseract.image_to_string(thresh,config = r"--psm 11",lang='eng+osd') #page segmentation mode 11 in config
print(extractedInformation)
```
